
# Multiple Regression: BA & AAPL vs. DJI
**Goal:** Reproduce and extend a classic finance example: how much of the Dow Jones Industrial Average (DJI) variation can be explained by Boeing (BA) and Apple (AAPL) stock prices?

This notebook downloads daily data, aligns it, and estimates a log-level model:

\begin{equation}
\ln(\text{DJI}_t) = \beta_0 + \beta_1 \ln(\text{BA}_t) + \beta_2 \ln(\text{AAPL}_t) + \varepsilon_t.
\end{equation}

We will also:
- Inspect the regression summary (coefficients, $R^2$, p-values)
- Run diagnostics (residual plots, Q-Q plot, heteroskedasticity test)
- Check multicollinearity (VIF)
- Try a robustness variant on **log returns** and with **Newey–West** standard errors

````python

import pandas as pd
import numpy as np
import yfinance as yf
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.stats.stattools import durbin_watson
from statsmodels.stats.outliers_influence import variance_inflation_factor
from scipy import stats

plt.rcParams["figure.figsize"] = (8, 5)
````

## 1) Download & prepare data

````python

import yfinance as yf

tickers = ["BA", "AAPL", "^DJI"]
start, end = "2015-01-01", None

# 方式一：直接取 Close
raw = yf.download(tickers, start=start, end=end)["Close"]

# 方式二：如果你真的想要 Adj Close，就強制 auto_adjust=False
# raw = yf.download(tickers, start=start, end=end, auto_adjust=False)["Adj Close"]

raw = raw.dropna(how="all")
raw.head()
````

````python

df = raw.copy().dropna()
df.columns = ["AAPL" if c=="AAPL" else "BA" if c=="BA" else "DJI" for c in df.columns]

# Log-levels
log_df = np.log(df[["DJI", "BA", "AAPL"]]).dropna()
log_df.head()
````

## 2) OLS on log-levels: ln(DJI) ~ ln(BA) + ln(AAPL)

````python

y = log_df["DJI"]
X = log_df[["BA", "AAPL"]]
X = sm.add_constant(X)

ols = sm.OLS(y, X).fit()
print(ols.summary())
````

## 3) Diagnostics

````python

resid = ols.resid
fitted = ols.fittedvalues

fig = plt.figure()
plt.scatter(fitted, resid, s=8)
plt.axhline(0, lw=1)
plt.xlabel("Fitted ln(DJI)")
plt.ylabel("Residuals")
plt.title("Residuals vs Fitted")
plt.show()

fig = plt.figure()
sm.qqplot(resid, line="s")
plt.title("Q-Q Plot of Residuals")
plt.show()

dw = durbin_watson(resid)
dw
````

````python

# Breusch–Pagan test for heteroskedasticity
bp_test = het_breuschpagan(resid, ols.model.exog)
bp_labels = ["Lagrange multiplier statistic", "p-value", "f-value", "f p-value"]
dict(zip(bp_labels, bp_test))
````

### Multicollinearity (VIF)

````python

vif_df = pd.DataFrame({
    "feature": ["const", "ln(BA)", "ln(AAPL)"],
    "VIF": [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
})
vif_df
````

## 4) Robustness: log returns & Newey–West (HAC)

````python

ret = np.log(df).diff().dropna()
y_r = ret["DJI"]
X_r = ret[["BA", "AAPL"]]
X_r = sm.add_constant(X_r)

ols_ret = sm.OLS(y_r, X_r).fit(cov_type="HAC", cov_kwds={"maxlags":5})
print(ols_ret.summary())
````


## 5) Interpretation & Notes

- In the **log-level** model, the coefficients can be read as elasticities: a 1% change in BA (or AAPL) price is associated with approximately $\beta$% change in DJI, holding the other constant.
- A very high $R^2$ on price levels can reflect **common trends**. This is why we also check **log-returns** and (optionally) **cointegration** for a more rigorous time-series treatment.
- If heteroskedasticity is present (BP test p-value small), consider using **robust** or **HAC/Newey–West** errors (we did for returns).
- If VIF is large (>10), multicollinearity may inflate standard errors; returns often mitigate this.
- For full time-series rigor, consider:
  - Unit-root tests (ADF) on levels
  - Cointegration tests (Johansen/Engle–Granger)
  - Error-correction models (ECM)
